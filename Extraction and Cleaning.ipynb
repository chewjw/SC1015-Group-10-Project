```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
# Reading in CSV file obtained from kaggle
df = pd.read_csv('gsm.csv')
df.head()
```

```python
# Extracting the top 15 phone brands
df15 = df[(df['oem'] =='Samsung') | (df['oem'] == 'Apple') | (df['oem'] == 'Huawei') | (df['oem'] == 'Xiaomi') | (df['oem'] == 'Oppo') | (df['oem'] == 'vivo') | (df['oem'] == 'Asus') | (df['oem'] == 'Microsoft') | (df['oem'] == 'Nokia') | (df['oem'] == 'LG') | (df['oem'] == 'Motorola') | (df['oem'] == 'Lenovo') | (df['oem'] == 'Sony') | (df['oem'] == 'Realme') | (df['oem'] == 'Google')]

# Extracting relevant variables
df15 = pd.DataFrame(df15[["oem","battery_charging","comms_gps","platform_cpu","platform_gpu","display","body_sim","network_technology","display_type","selfie_camera_features","display_protection", "model", "launch_status", "main_camera_video", "selfie_camera_video",  "body_weight", "display_resolution", "display_size", "sound_3.5mm_jack", "memory_internal", "platform_os", "platform_chipset", "battery", "misc_price"]])

# Remove discontinued and cancelled
df15 = df15[(df15['launch_status'] != "Discontinued") & (df15['launch_status'] != "Cancelled")]
df15 = df15.drop('launch_status', 1)
df15.head()
```

# 2. Data Cleaning

```python
# Cleaning body weight
df15['body_weight'] = df15['body_weight'].astype("string")
j = []
df15['body_weight'].fillna(value = '0', inplace = True)
for i in df15['body_weight']:
    if (i):
        i = i.split('g')[0]
        i = i.split(' ')[0]
        if (i == '-') | (i == 'V1'):
            i = '0'
    else:
        i = 0
    j.append(i)
df15['body_weight'] = j
```

```python
# clean display_resolution (width x height)
df15['display_resolution'] = df15['display_resolution'].astype("string")
j = []
df15['display_resolution'].fillna(value = 'unknown', inplace = True)
for i in df15['display_resolution']:
    i = i.split('p')[0]
    j.append(i)    
df15['display_resolution'] = j
```

```python
# clean display_size (inches)
df15['display_size'] = df15['display_size'].astype("string")
j = []
df15['display_size'].fillna(value = 'unknown', inplace = True)
for i in df15['display_size']:
    i = i.split(' ')[0]
    i = i.split('i')[0]
    j.append(i)   
df15['display_size'] = j
```

```python
#clean platform_os
df15['platform_os'] = df15['platform_os'].astype("string")
j = []
df15['platform_os'].fillna(value = 'unknown', inplace = True)
for i in df15['platform_os']:
    i = i.split(' ')[0]
    j.append(i)      
df15['platform_os'] = j
```

```python
#clean memory_internal
df15['memory_internal'] = df15['memory_internal'].astype("string")
j = []
for i in df15['memory_internal']:
    i = i.split('RAM,')[0]
    i = i.split('RAM')[0]
    j.append(i)    
df15['memory_internal'] = j
x = df15['memory_internal'].unique()
```

```python
#splitting memory internal
df15['memory_internal'] = df15['memory_internal'].astype("string")
df15[['memory','ram']] = df15.memory_internal.str.split(n=1,expand=True)
```

```python
#clean memory
df15['memory'] = df15['memory'].astype("string")
j = []
for i in df15['memory']:
    i = i.split(' ')[0]
    i = i.split('/')[0]
    i = i.split('G')[0]
    j.append(i)       
df15['memory'] = j
```

```python
# Clean ram
df15['ram'] = df15['ram'].astype("string")
j = []
for i in df15['ram']:
    i = i.split('GB,')[-1]
    i = i.split('GB')[0]
    i = i.split('/')[0]
    j.append(i)    
df15['ram'] = j

# Further Cleaning
for i in df15['ram']:
    df15['ram'] = df15['ram'].astype("string")
    df15['ram'] = df15['ram'].replace(' 512 MB ', '0.5')
    df15['ram'] = df15['ram'].replace('TB, 12 ', '12')
    df15['ram'] = df15['ram'].replace('0.5 GB', '0.5')
    
df15.loc[ df['model'] == 'X max', 'ram'] = '1.5'
df15.loc[ df['model'] == 'Galaxy Note 3 Neo Duos', 'ram'] = '2'
df15.loc[ df['model'] == 'Galaxy Note 3 Neo', 'ram'] = '2'
```

```python
# Clean price
df15['misc_price'] = df15['misc_price'].astype("string")
j = []
j2 = []
df15['misc_price'].fillna(value = '0', inplace = True)
for i in df15['misc_price']:
    x = '<c2><a3><e2><80><89>'
    y = '<e2><82><ac><e2><80><89'
    z = '<e2><80><89>'

    l = []
    if x in i:
        k = i.split('<c2><a3><e2><80><89>')[-1]
        l.append(k.split(' / ')[0])
    if y in i:
        k = i.split('<e2><82><ac><e2><80><89>')[-1]
        l.append(k.split(' / ')[0])
    if z in i:
        k = i.split('<e2><80><89>')[-1]
        l.append(k.split(' / ')[0])
    if (l == []):
        if (i == '0'):
            l.append('0')
        else:
            k = i.split(' ')[1]
            l.append(k)

    min_price = float(l[0].replace(",", ""))
    
    for num in range(len(l)):
        y = float(l[num].replace(",", ""))
        if y < min_price:
            min_price = y
    min_price = round(min_price, 2)
    j.append(min_price)
df15['misc_price'] = j
```

```python
# Split battery into removability and battery_capacity
df15['battery'] = df15['battery'].astype("string")
df15[['removability','battery_capacity']] = df15.battery.str.split(n=1,expand=True)
```
